# LEVEL 1 DATA

## DSCOVR data

```python
import xarray as xr
import os

import gzip

nc_files = []

for file in os.listdir('compressed'):
    output_file = os.path.join('uncompressed/',file[:-3])  # The decompressed file
    file = os.path.join('compressed/',file)
    nc_files.append(output_file)
    with gzip.open(file, 'rb') as compressed_file:
        with open(output_file, 'wb') as decompressed_file:
            decompressed_file.write(compressed_file.read())

```

## FARADAY CUP

```python
nc_file = nc_files[0]

dataset = xr.open_dataset(nc_file)

# Convert the entire dataset into a Pandas DataFrame
df = dataset.to_dataframe()

#create list with proton measurements

important_variables = []

#due to the low relevance in solar wind prediction

for variable in df.columns.values:
    if 'proton' in variable:
        important_variables.append(variable)

#show the proton measurement

print(important_variables)
```

```python
important_variables = ['proton_vx_gse', 'proton_vy_gse', 'proton_vz_gse', 'proton_vx_gsm', 'proton_vy_gsm', 'proton_vz_gsm', 'proton_speed', 'proton_density', 'proton_temperature']

faraday_cup = df[important_variables]

faraday_cup = faraday_cup.resample('1min').mean()
```

## MAGNETOMETER

```python
nc_file = nc_files[1]  # Replace with the actual .nc file name

dataset = xr.open_dataset(nc_file)
# Convert the entire dataset into a Pandas DataFrame
df = dataset.to_dataframe()

important_variables = ['bt','bx_gse', 'by_gse', 'bz_gse', 'theta_gse', 'phi_gse', 'bx_gsm','by_gsm', 'bz_gsm', 'theta_gsm', 'phi_gsm']


magnetometer = df[important_variables]
# Print the first few rows of the DataFrame

magnetometer = magnetometer.resample('1min').mean()
```
